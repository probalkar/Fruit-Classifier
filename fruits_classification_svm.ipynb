```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
fruits['fruit_label']
```

```python
# fruits.head
features = ['mass', 'width', 'height', 'color_score']
X = fruits[features]
y = fruits['fruit_label']
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=49, shuffle=True)
```

```python
X_train
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.svm import SVC
```

```python
svm = SVC()
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
svm.fit(X_train, y_train)
```

```python
print("Accuracy on training set: ", svm.score(X_train, y_train))
print("Accuracy on testing set: ", svm.score(X_test, y_test))
```

```python
y_pred = svm.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
print("Acuracy score: ", accuracy_score(y_test, y_pred))
```
